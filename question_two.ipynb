Here’s a Python code snippet that performs hierarchical clustering on the given distance matrix and draws its dendrogram. We’ll use the scipy library for this purpose.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Given distance matrix
distance_matrix = [
    [0, 1, 4, 5],
    [1, 0, 2, 6],
    [4, 2, 0, 3],
    [5, 6, 3, 0]
]

# Perform hierarchical clustering (agglomerative)
Z = linkage(distance_matrix, 'single')

# Draw dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=['A', 'B', 'C', 'D'])
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

This code will create a dendrogram that represents the hierarchical clustering of the data points A to D based on the provided distance matrix. The dendrogram is a tree-like diagram that shows the order in which clusters were merged.
